```python
import pandas as pd

df = pd.read_csv("salaries_by_college_major.csv")
df.head()
```

```python
clean_df = df.dropna()
clean_df.tail()
```

```python
# max Starting Median Salary and its major
max_starting_median_salary = clean_df["Starting Median Salary"].max()

index_of_max_starting_median_salary = clean_df["Starting Median Salary"].idxmax()

major_of_max_starting_median_salary = clean_df["Undergraduate Major"].loc[index_of_max_starting_median_salary]

major_of_max_starting_median_salary, max_starting_median_salary
```

```python
# max Mid-Career Median Salary and its major
max_mid_career_median_salary = clean_df["Mid-Career Median Salary"].max()

index_of_max_mid_career_median_salary = clean_df["Mid-Career Median Salary"].idxmax()

major_of_max_mid_career_median_salary = clean_df["Undergraduate Major"][index_of_max_mid_career_median_salary]

major_of_max_mid_career_median_salary, max_mid_career_median_salary
```

```python
# min Starting Median Salary and its major
min_starting_median_salary = clean_df["Starting Median Salary"].min()

index_of_min_starting_median_salary = clean_df["Starting Median Salary"].idxmin()

major_of_min_starting_median_salary = clean_df["Undergraduate Major"].loc[index_of_min_starting_median_salary]

major_of_min_starting_median_salary, min_starting_median_salary
```

```python
# min Mid-Career Median Salary and its major
min_mid_career_median_salary = clean_df["Mid-Career Median Salary"].min()

index_of_min_mid_career_median_salary = clean_df["Mid-Career Median Salary"].idxmin()

major_of_min_mid_career_median_salary = clean_df["Undergraduate Major"][index_of_min_mid_career_median_salary]

major_of_min_mid_career_median_salary, min_mid_career_median_salary
```

```python
# Low-risk major has the minimum difference between the 10th and 90th percentile:
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, "Spread", spread_col)
clean_df.head()
```

```python
low_risk = clean_df.sort_values("Spread")
low_risk[["Undergraduate Major", "Spread"]].head()
```

```python
# top 5 degrees with the highest values in the 90th percentile.
clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending=False)[["Undergraduate Major", "Mid-Career 90th Percentile Salary"]].head()
```

```python
# the degrees with the greatest spread in salaries
high_risk = clean_df.sort_values("Spread", ascending=False)
high_risk[["Undergraduate Major", "Spread"]].head()
```

```python
# top 5 degrees with the highest values in the median salaries.
clean_df.sort_values("Mid-Career Median Salary", ascending=False)[["Undergraduate Major", "Mid-Career Median Salary"]].head()
```

```python
# how many majors we have in each group:
clean_df.groupby("Group").count()
```

```python
# what are the average salaries per group:
clean_df.groupby("Group").mean(numeric_only=True)
```

```python
# round 2 for better redability
pd.options.display.float_format = '{:,.2f}'.format
clean_df.groupby("Group").mean(numeric_only=True)
```
